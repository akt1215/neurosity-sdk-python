```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('/workspaces/neurosity-sdk-python/base2.csv')
grab = pd.read_csv('/workspaces/neurosity-sdk-python/grab2.csv')
```

```python
base = (2/(1+1.01**base)-1)*50
grab = (2/(1+1.01**grab)-1)*50
```

```python
# add values to the "target" column. The number should be 0
base['target'] = 0
grab['target'] = 1
```

```python
# plot a histogram of the data
# make the size bigger
plt.rcParams['figure.figsize'] = [20, 10]
base.hist()
```

```python
# plot a histogram of the data
# make the size bigger
plt.rcParams['figure.figsize'] = [20, 10]
grab.hist()
```

```python
#count the number of columns with bigger than 10000
(grab > 100).sum()
#(data < -1000).sum()
```

```python
# Merge the data
data = pd.concat([base, grab], axis=0)
#shuffle the data
data = data.sample(frac=1)
data.describe()
```

```python
data.describe()
```

```python
data.describe()
```

```python
cor = data.corr()
plt.figure(figsize=(15,10))
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
```

```python
import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('target', axis=1), data['target'], test_size=0.2, random_state=42)

model = Sequential()
model.add(Dense(64, input_dim = X_train.shape[1], activation='relu'))
#model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

number_of_epochs = 30
model.fit(X_train, y_train, epochs=number_of_epochs, batch_size=5, validation_split=0.2)

plt.scatter(range(number_of_epochs), model.history.history['accuracy'], s=30)
predictions = model.predict(X_test)
print("Accuracy: ", model.evaluate(X_test, y_test)[1])
print(predictions[:5])
print(y_test[:5])
```
