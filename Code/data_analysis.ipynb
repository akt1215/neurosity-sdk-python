```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/workspaces/neurosity-sdk-python/Datas/Aliceblue mole.csv')
data.describe()
```

```python
data["A"] = 100*data["A"]/data["A"].max()
data["B"] = 100*data["B"]/data["B"].min()
data["C"] = 100*data["C"]/data["C"].min()
data["D"] = 100*data["D"]/data["D"].min()
data["E"] = 100*data["E"]/data["E"].min()
data["F"] = 100*data["F"]/data["F"].min()
data["G"] = 100*data["G"]/data["G"].min()
data["H"] = 100*data["H"]/data["H"].min()
data["I"] = 100*data["I"]/data["I"].min()
data.describe()
```

```python
plt.figure(figsize=(100,50))
plt.plot(data["A"], label="A")
plt.plot(data["C"], label="C")
plt.plot(data["D"], label="D")
plt.plot(data["E"], label="E")
plt.plot(data["G"], label="G")
plt.plot(data["H"], label="H")

plt.legend()
plt.show()
```
